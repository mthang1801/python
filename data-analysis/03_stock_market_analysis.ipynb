### Tỉ lệ thay đổi hàng ngày (Daily percent change) 

Đầu tiên, chúng ta sẽ bắt đầu bằng cách tính toán phần trăm thay đổi hàng ngày. Phần trăm thay đổi hàng ngày được xác định theo công thức sau:

$r_t = \frac{p_t}{p_{t-1}} - 1$

Trong đó : 
+ $r_t$ : Tỉ lệ giá cổ phiếu tại thời điểm t so với thời điểm t-1. Để biết thời điểm t này có lời hay lỗ so với thời điểm trước, chỉ cần trừ đi 1. Nếu tỉ lệ này là dương nghĩa là giá cổ phiếu thời điểm t cao hơn thời điểm t-1, và ngược lại. 
+ $p_t$ : Giá cổ phiếu thời điểm t 
+ $p_{t-1}$ : Giá cổ phiếu thời điểm t-1

Trong Python, chúng ta sẽ áp dụng phương thức `shift()` trong pandas để lấy dữ liệu ở thời điểm trước và sau đó thực hiện công thức trên.



```python
pfe["Daily Change"] = pfe["Close"] / pfe["Close"].shift(1) - 1
jnj["Daily Change"] = jnj["Close"] / jnj["Close"].shift(1) - 1
azn["Daily Change"] = azn["Close"] / azn["Close"].shift(1) - 1
```

```python
pfe["Daily Change"].hist(figsize=(12,7),label="PFE", alpha=0.4,bins=50)
jnj["Daily Change"].hist(label="JNJ",  alpha=0.4, bins=50)
azn["Daily Change"].hist(label="AZN",  alpha=0.4, bins=50)
plt.title("Tỉ lệ cổ phiếu thay đổi hàng ngày")
plt.legend()
```

```python
pfe["Daily Change"].plot.kde(figsize=(12,7),label="PFE")
jnj["Daily Change"].plot.kde(label="JNJ")
azn["Daily Change"].plot.kde(label="AZN")
plt.title("Tỉ lệ cổ phiếu thay đổi hàng ngày")
plt.legend()
```

Biểu đồ nào có đường cong càng nhô cao hoặc càng nhọn thì tỉ lệ thay đổi càng thấp. Ngược lại, biểu đồ nào càng thấp hoặc càng phình ra 2 bên thì tỉ lệ thay đổi càng cao. Qua đây, có thể thấy JNJ có tỉ lệ thay đổi hàng ngày thấp, cho thấy giá trị cổ phiếu của công ty này tương đối ổn định. Trong khi Astrazeneca có tỉ lệ thay đổi cao, giá cổ phiếu tăng giảm liên tục.

```python
fig, axes = plt.subplots(1,3,figsize=(20,5))

axes[0].hist(pfe["Daily Change"],bins=50)
axes[0].set_title("PFE")

axes[1].hist(jnj["Daily Change"],bins=50)
axes[1].set_title("JNJ")

axes[2].hist(azn["Daily Change"],bins=50)
axes[2].set_title("AZN")
plt.suptitle("Biểu đồ tỉ lệ thay đổi giá cổ phiếu hàng ngày")
```

### Lợi nhuận tích lũy hàng ngày (Cumulative Daily Returns) 

Tỉ lệ thay đổi hàng ngày chỉ cho chúng ta biết được giá cổ phiếu của ngày hôm đó so với ngày hôm trước tăng hay giảm, nó chỉ phản ánh thực trạng giá cổ phiếu trong ngắn hạn để nhà đầu tư có thể nhìn nhận các tình huống có thể xảy ra trong tương lai gần, ngoài ra nó không phản ánh được bất kỳ điều gì. 

Ngoài việc quan tâm đến giá cổ phiếu thay đổi hàng ngày như thế nào, nhà đầu tư cũng thường quan tâm đến lợi nhuận tích lũy sau một khoảng thời gian kể từ ngày bắt đầu đầu tư cổ phiếu. VD: Mua 1 cổ phiếu ABC vào ngày 1-1-2021 theo bảng giá sau: 

    Date                        Price
    01/01/2021                   10
    02/01/2021                   15
    03/01/2021                   20
    04/01/2021                   25

**Lợi nhuận hàng ngày (Daily Return)** : Là khoản lãi/lỗ về giá cổ phiếu so với ngày trước. Nếu giá trị này là dương nghĩa là có lời, nếu giá cổ phiếu giảm nghĩa là lỗ. Nó cũng được thể hiện theo tỷ lệ phần trăm để truyền tải thông tin tốt hơn (Như phần trên). Vì vậy, đối với ví dụ trên, lợi nhuận hàng ngày sẽ là : 

    Date                         Daily Return                  %Daily Return
    01/01/2018                 10/10 =  1                          -   
    01/02/2018                 15/10 =  3/2                       50%
    01/03/2018                 20/15 =  4/3                       33%
    01/04/2018                 25/20 =  5/4                       20%


**Lợi nhuận tích lũy** : Mặc dù lợi nhuận hàng ngày rất hữu ích, nhưng nó không cung cấp cho nhà đầu tư cái nhìn sâu sắc ngay lập tức về lợi nhuận mà anh ta đã đạt được cho đến nay, đặc biệt nếu giá cổ phiếu rất dễ biến động. Lợi nhuận tích lũy được tính tương ứng với ngày bắt đầu đầu tư. Nếu lợi nhuận tích lũy >0 , bạn đang tạo ra lợi nhuận, ngược lại bạn đang thua lỗ. Vì vậy, đối với ví dụ trên, lợi nhuận tích lũy như sau

    Date                       Cumulative Return         %Cumulative Return
    01/01/2018                  10/10 =  1                         100 %   
    01/02/2018                  15/10 =  3/2                       150 %
    01/03/2018                  20/10 =  2                         200 %
    01/04/2018                  25/10 =  5/2                       250 %

  
Công thức lợi nhuận tích lũy : 

$i_t = \frac{p_t}{p_0}$

hoặc 

$ i_i = (1+r_t) * i_{t-1} $ 

Pandas làm cho điều này rất đơn giản để tính toán với phương thức `cumprod()` của nó. Sử dụng theo cách sau:


    df[daily_cumulative_return] = ( 1 + df[pct_daily_return] ).cumprod()






```python
# VD
df = pd.DataFrame({"A" : [1,2,3,4,5]})
df["A"].cumsum()
```

```python
pfe["Cummulative Return"] = (1 + pfe["Daily Change"]).cumprod()
jnj["Cummulative Return"] = (1 + jnj["Daily Change"]).cumprod()
azn["Cummulative Return"] = (1 + azn["Daily Change"]).cumprod()
```

```python
pfe["Cummulative Return"].plot(figsize=(12,7), label="PFE")
jnj["Cummulative Return"].plot(label="JNJ")
azn["Cummulative Return"].plot(label="AZN")
plt.title("Tỉ lệ lợi nhuận tích lũy")
plt.legend()
```
