## Matplotlib cơ bản

Trước tiên, chúng ta sẽ import một số thư viện liên quan: 
+ `numpy` : tạo mảng dữ liệu 
+ `pandas` : cấu trúc các dữ liệu

```python
import numpy as np 
import pandas  as pd 
import matplotlib.pyplot as plt 
```

### Tạo biểu đồ đơn

Chúng ta có thể tạo một biểu đồ dòng rất đơn giản bằng cách sử dụng cách sau (bạn có thể mở gợi ý bằng Shift + Tab để kiểm tra tài liệu sử cho các chức năng chúng ta đang sử dụng).


```python
# Tạo dữ liệu mẫu
x = np.linspace(0,5,10)
y = x ** 2 
```

```python
plt.plot(x,y)
plt.xlabel("Trục X")
plt.ylabel("Truc y")
```

### Tạo nhiều biểu đồ trên cùng canvas

```python
plt.subplot(1,2,1) # biểu đồ gồm 1 dòng 2 cột và vị trí số 1
plt.plot(x,y,".-")
plt.subplot(1,2,2) # biểu đồ gồm 1 dòng 2 cột và vị trí số 2
plt.plot(y,x, "g*-")
```

### Phương pháp hướng đối tượng trong Matplotlib

Bây giờ chúng ta đã thấy biểu đồ cơ bản, tiếp theo chúng ta sẽ đi vào chi tiết về các API hướng đối tượng trong matplotlib. Ý tưởng chính trong việc sử dụng phương pháp Hướng đối tượng là tạo một object figure sau đó, sử dụng các phương thực và thuộc tính của nó để biểu diễn đồ thị. Cách tiếp cận này tốt hơn khi xử lý một canvas có nhiều biểu đồ trên đó.

Để bắt đầu, chúng ta tạo một figure. Sau đó, chúng ta có thể thêm các trục vào hình đó:



### plt.figure()

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])

axes.plot(x,y, "*-")
axes.set_title("Phương pháp hướng đối tượng trong Matplotlib")
axes.set_xlabel("x")
axes.set_ylabel("y")
```

Khi có nhiều biểu đồ trong môt canvas, chúng ta có thể thấy được sự tiện lợi khi sử dụng phương pháp hướng đối tượng này. Ví dụ cho biểu đồ phức tạp hơn : 

```python
fig = plt.figure()
# Sử dụng phương thức add_axes để thêm một khung hình 
axes1 = fig.add_axes([0,0,1,1]) # mảng gồm 4 phần tử [cách biên trái, cách biên dưới, chiều rộng khung, chiều cao khung]
axes2 = fig.add_axes([0.1,0.5,0.4,0.4])
axes1.plot(x,y,"--")
axes1.set_xlabel("X axes1")
axes1.set_ylabel("y axes1")

axes2.plot(y,x,"g-*")
axes1.set_xlabel("X axes2")
axes1.set_ylabel("y axes2")
```

### Legends, labels and titles

Những kiến ​​thức cơ bản về cách tạo một canvas hình và thêm các thể hiện trục vào canvas đã xong, hãy xem cách trang trí một hình với tiêu đề, label trên trục và chú thích của một biểu đồ.

**Sử dụng phương pháp thông thường :**

```python
plt.plot(x,y, "g-+", label="$x^2$") # cần phải đặt label để hiện thị chú thích
plt.plot(x,x**3, "r.-", label="$x^3$")  
plt.title("Biểu đồ x và x^2") # Hiển thị tiêu đề của biểu đồ
plt.xlabel("x") # Hiển thị label của trục x 
plt.ylabel("$x^2$") # Hiển thị label của trục y 
plt.legend() # Hiển thị chú thích (nếu không đặt thì sẽ không hiển thị chú thích)
```

**Sử dụng phương pháp hướng đối tượng :**

Phương pháp này có một chút khác biệt so với phương pháp vẽ biểu đồ thông thường, chẳng hạn `plt.title()` thì `axes.set_title()`... 

```python
fig, axes = plt.subplots()

axes.plot(x,y, "g-+", label="$x^2$") # cần phải đặt label để hiện thị chú thích
axes.plot(x,x**3, "r.-", label="$x^3$")  
axes.set_title("Biểu đồ x và x^2") # Hiển thị tiêu đề của biểu đồ
axes.set_xlabel("x") # Hiển thị label của trục x 
axes.set_ylabel("$x^2$") # Hiển thị label của trục y 
axes.legend() # Hiển thị chú thích (nếu không đặt thì sẽ không hiển thị chú thích)
```

Lưu ý rằng ô chú thích có thể chồng lên các đường biểu diễn trong biểu đồ, chúng ta có thể thay đổi vị trí ô chú thích tùy ý với tham số `loc` trong `axes.legend()`. Xem thêm tại : https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.legend.html


```python
fig, axes = plt.subplots()

axes.plot(x,y, "g-+", label="$x^2$") # cần phải đặt label để hiện thị chú thích
axes.plot(x,x**3, "r.-", label="$x^3$")  
axes.set_title("Biểu đồ x và x^2") # Hiển thị tiêu đề của biểu đồ
axes.set_xlabel("x") # Hiển thị label của trục x 
axes.set_ylabel("$x^2$") # Hiển thị label của trục y 
axes.legend(loc="upper center") # Hiển thị chú thích (nếu không đặt thì sẽ không hiển thị chú thích)
```

### Thiết lập màu sắc cho các đường biểu đồ 

Với matplotlib, chúng ta có thể xác định màu sắc của các đường và các yếu tố đồ họa khác theo một số cách. Trước hết, chúng ta có thể sử dụng cú pháp giống MATLAB trong đó 'b' có nghĩa là màu xanh lam, 'g' có nghĩa là màu xanh lá cây, v.v. API MATLAB để chọn kiểu dòng cũng được hỗ trợ: ví dụ: trong đó, 'b.-' có nghĩa là a đường màu xanh lam với các dấu chấm:

```python
# MATLAB style line color and style 
fig, ax = plt.subplots()
ax.plot(x, 
        x ** 2, 
        'b.-') # blue line with dots
ax.plot(x, 
        x ** 3, 
        'g--') # green dashed line
```

Chúng ta cũng có thể xác định màu theo tên của chúng hoặc mã hệ lục phân RGB và tùy chọn cung cấp giá trị alpha bằng cách sử dụng các đối số từ khóa màu và alpha. Alpha cho biết độ mờ.

```python
fig, ax = plt.subplots()

ax.plot(x, x + 1, color = "blue", alpha=0.5) # half-transparant
ax.plot(x, x + 2, color = "#8B008B")        # RGB hex code
ax.plot(x, x + 3, color = "#FF8C00")        # RGB hex code 
```

### Kiểu đường kẻ và điểm đánh dấu

Để thay đổi độ rộng của đường biểu diễn, chúng ta có thể sử dụng đối số từ khóa `linewidth` hoặc `lw`. Kiểu đường biểu diễn có thể được chọn bằng cách sử dụng các đối số từ khóa `linestyle` hoặc `ls`:




```python
fig, ax = plt.subplots(figsize=(12,6))

ax.plot(x, x + 1, color = "red", linewidth = 0.25)
ax.plot(x, x + 2, color = "red", linewidth = 0.50)
ax.plot(x, x + 3, color = "red", linewidth = 1.00)
ax.plot(x, x + 4, color = "red", linewidth = 2.00)

# possible linestype options ‘-‘, ‘–’, ‘-.’, ‘:’, ‘steps’
ax.plot(x, x + 5, color = "green", lw = 3, linestyle = '-')
ax.plot(x, x + 6, color = "green", lw = 3, ls = '-.')
ax.plot(x, x + 7, color = "green", lw = 3, ls = ':')

# custom dash
line, = ax.plot(x, x + 8, color = "black", lw = 1.50)
line.set_dashes([5, 10, 15, 10]) # format: line length, space length, ...

# possible marker symbols: marker = '+', 'o', '*', 's', ',', '.', '1', '2', '3', '4', ...
ax.plot(x, x + 9, color = "blue", lw = 3, ls = '-', marker = '+')
ax.plot(x, x + 10, color = "blue", lw = 3, ls = '--', marker = 'o')
ax.plot(x, x + 11, color = "blue", lw = 3, ls = '-', marker = 's')
ax.plot(x, x + 12, color = "blue", lw = 3, ls = '--', marker = '1')

# marker size and color
ax.plot(x, x + 13, color = "purple", lw = 1, ls = '-', marker = 'o', markersize = 2)
ax.plot(x, x + 14, color = "purple", lw = 1, ls = '-', marker = 'o', markersize = 4)
ax.plot(x, x + 15, color = "purple", lw = 1, ls = '-', marker = 'o', markersize = 8, markerfacecolor = "red")
ax.plot(x, x + 16, color = "purple", lw = 1, ls = '-', marker = 's', markersize = 8, 
        markerfacecolor = "yellow", markeredgewidth = 3, markeredgecolor = "green");
```

### Kiểm soát giới hạn của trục

Chúng ta có thể thiết lập phạm vi của các trục bằng cách sử dụng phương thức `set_ylim` và `set_xlim`:


```python
fig, axes = plt.subplots(1, 3, figsize = (12, 4))

axes[0].plot(x, x ** 2, x, x ** 3)
axes[0].set_title("default axes ranges")

axes[1].plot(x, x ** 2, x, x ** 3)
axes[1].axis('tight')
axes[1].set_title("tight axes")

axes[2].plot(x, x ** 2, x, x ** 3)
axes[2].set_ylim([0, 60])
axes[2].set_xlim([2, 5])
axes[2].set_title("custom axes range");
```

### Logarithmic scale

Bạn có thể đặt thang đo logarit cho một hoặc cả hai trục. Tính năng này trên thực tế chỉ là một ứng dụng của hệ thống biến đổi tổng quát hơn trong Matplotlib. Mỗi tỷ lệ của trục được đặt riêng biệt bằng cách sử dụng các phương thức `set_xscale` và `set_yscale` chấp nhận một tham số (với giá trị "log" trong trường hợp này):


```python
fig, axes = plt.subplots(1,2,figsize=(12,5))
axes[0].plot(x, x**2)
axes[0].plot(x, x**3)
axes[0].set_title("Normal scale")

axes[1].plot(x, x**2)
axes[1].plot(x, x**3)
axes[1].set_yscale("log")
axes[1].set_title("Log scale")
```

### Thiết lập vị trí của ticks và label cho ticks

```python
fig, ax = plt.subplots(figsize=(12,5))

ax.plot(x, x**2)
ax.plot(x, x**3)
ax.set_xticks(range(6))
ax.set_xticklabels([r'$\alpha$', r'$\beta$', r'$\gamma$', r'$\delta$', r'$\epsilon$'], 
                   fontsize = 18)
y_ticks=[0,50,100,150]
ax.set_yticks(y_ticks)
ax.set_yticklabels(["$%.2f$" % y for y in y_ticks])
```

### Số trục và khoảng cách label với trục


```python
import matplotlib

fig, ax = plt.subplots()

# khoảng cách giữa trục x và y và các số trên trục
matplotlib.rcParams["xtick.major.pad"] = 5
matplotlib.rcParams["ytick.major.pad"] = 5

ax.plot(x,x**2, x, x**3)
ax.set_xlabel("x", fontsize=14)
ax.set_ylabel("y", fontsize=14)
# padding giữa label và số trục
ax.xaxis.labelpad = 5
ax.yaxis.labelpad = 5
```

### Twin axes

Đôi khi rất hữu ích khi có trục x hoặc y kép trong một hình; ví dụ, khi vẽ đồ thị các đường cong với các đơn vị khác nhau với nhau. Matplotlib hỗ trợ điều này với các hàm `twinx` và `twiny`:


```python
fig, ax1 = plt.subplots()

ax1.plot(x,x**2, "b--")
ax1.set_ylabel("area ($m^2$)", color="blue")

for label in ax1.get_yticklabels(): 
  label.set_color("blue")


ax2 = ax1.twinx()
ax2.plot(x,x**3,"-g+")
ax2.set_ylabel("volume ($m^3$)", color="green")
for label in ax2.get_yticklabels() : 
  label.set_color("green")
```

### Các trục trong đó x và y bằng 0


```python
data = np.linspace(-1,1,100)
fig, ax = plt.subplots()

ax.plot(data,data**3)
ax.spines["left"].set_position(("data", 0))  # set position of x spine to x=0
ax.spines["bottom"].set_position(("data",0))  # set position of y spine to y=0

ax.set_xticks([-1,-0.5,0,0.5,1])
ax.set_yticks([-1,-0.5,0,0.5,1])
```

### Các kiểu trục đồ thị 2D khác

Ngoài phương pháp đồ thị đường thẳng thông thường, có một số kiểu đồ thị khác được sử dụng rất phổ biến trong các dự án:

```python
xx = np.random.rand(50)
yy = xx + 0.2 * np.random.randn(50)
fig, axes = plt.subplots(1,4,figsize=(25,5))

axes[0].scatter(xx,yy, s=50,edgecolor="black")
axes[0].set_title("scatter")

axes[1].step(x,x**2, lw=2, color="black")
axes[1].set_title("step")

axes[2].bar(x, x**2, width=0.4, alpha=0.7)
axes[2].set_title("bar")

axes[3].fill_between(x , x**2, x, x**3, color="green")
```

```python
x1 = np.linspace(-5,5,100)

fig, ax = plt.subplots()
ax.plot(x1, np.sin(x1), "--b")
ax.plot(x1, np.cos(x1), "-.r")
ax.axis("equal")
ax.text(-0.75,1.2, "$y = cos(x)$")
ax.text(-2, -1.4, "$y = sin(x)$")
```
